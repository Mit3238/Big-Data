```
%scala
// matrix multiplication
var a = Array(List(1,2,3), List(1,2,3), List(1,2,3))
var b = Array(List(4,5,6), List(4,5,6), List(4,5,6))
var c = Array(Array(0,0,0), Array(0,0,0), Array(0,0,0))
var sum = 0

for (i<-0 to 2) {
  for (j<-0 to 2) {
    sum = 0
    for (k<-0 to 2) {
      sum = sum + (a(i)(k) * b(k)(j))
    }
    c(i)(j) = sum
  }
}
println(c)
```
