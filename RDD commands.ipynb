```
%scala
def isPrime(num:Int):Boolean =
(num > 1) && !(2 to scala.math.sqrt(num).toInt).exists(x => num % x == 0)
```

```
%scala
val a = sc.parallelize(1 to 10)
val even = a.filter(x => (x % 2 == 0)).map(x => x*x).sum
val odd = a.filter(x => (x % 2 != 0)).map(x => x*x).sum
val prime = a.filter(isPrime).map(x => x*x).sum

println("Sum of odd is "+odd)
println("Sum of even is "+even)
println("Sum of prime is "+prime)
```

```
%scala
import org.apache.spark.sql.SparkSession
val spark = SparkSession.builder().appName("SparkExample").master("local").getOrCreate()
```

```
%scala
val state = Map(("NY", "New York"), ("CA", "California"), ("FL", "Florida"))
```

```
%scala
val contries = Map(("USA", "America"), ("IN", "India"))
```

```
%scala
val brodState = spark.sparkContext.broadcast(state)
```

```
%scala
val brodContries = spark.sparkContext.broadcast(contries)
```

```
%scala
val data = Seq(("Mit", "Patel", "IN", "CA"),
               ("Smit", "Patel", "USA", "CA"),
               ("Mire", "Patel", "IN", "NY"),
               ("Parth", "Patel", "USA", "FL"))
```

```
%scala
val rdd = spark.sparkContext.parallelize(data)
```

```
%scala
val rdd2 = rdd.map(f => {
  val country = f._3
  val state = f._4
  val fullCountry = brodContries.value.get(country).get
  val fullState = brodState.value.get(state).get
  (f._1, f._2, fullCountry, fullState)
})
```

```
%scala
println(rdd2.collect().mkString("\n"))
```

```
%scala
val state = Map(("NY", "New York"), ("CA", "California"), ("FL", "Florida"))
val contries = Map(("USA", "America"), ("IN", "India"))
val brodState = spark.sparkContext.broadcast(state)
val brodContries = spark.sparkContext.broadcast(contries)

val data = Seq(("Mit", "Patel", "IN", "CA"),
               ("Smit", "Patel", "USA", "CA"),
               ("Mire", "Patel", "IN", "NY"),
               ("Parth", "Patel", "USA", "FL"))
val columns = Seq("firstname", "lastname", "Country", "State")

import spark.sqlContext.implicits._

val df = data.toDF(columns:_*)
val df2 = df.map(row=>{
  val country = row.getString(2)
  val state = row.getString(3)
  val fullState = brodState.value.get(state).get
  val fullCountry = brodContries.value.get(country).get
  (row.getString(0), row.getString(1), fullCountry, fullState)
}).toDF(columns:_*)
```

```
%scala
df2.show(4)
```

```
%scala
def param0 = (acc:Int, v:Int) => acc + v
def param1 = (acc1:Int, acc2:Int) => acc1 + acc2
println("Aggregate : " + listRDD.aggregate(0) (param0, param1))
```

```
%scala
def param3 = (acc:Int, v:(String, Int)) => acc + v._2
def param2 = (acc1:Int, v2:Int) => acc1 + v2

println("Aggregate : " + inputRDD.aggregate(0) (param3, param2))
```

```
%scala
def param4 = (acc:Int, v:Int) => acc + v
def param5 = (acc1:Int, v2:Int) => acc1 + v2

println("Tree Aggregate : " + listRDD.treeAggregate(0) (param4, param5))
```
